## 2. Line plot <a class="anchor" id="chapter2"></a>

**Figure** (an instance of the class `plt.Figure`) is a single container that contains all the objects representing axes, graphics, text, and labels. 

**The axes** (an instance of the class `plt.Axes`) is what we see above: a bounding box with ticks and labels, which will eventually contain the plot elements that make up our visualization.

```python
plt.style.use('seaborn-whitegrid')
figure(figsize=(4, 3), dpi=80)
ax = plt.axes()
x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x))
```

### Adjust the plot - Line, Color and Styles <a class="anchor" id="chapter2.1"></a>

* Use `color` keyword to specify, can be in many ways such as:

```python
figure(figsize=(4, 3), dpi=80)

# Commonly use:
plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)

# Others
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

If no color is specified, Matplotlib will automatically cycle through a set of default colors for multiple lines.



* Similarly, the line style can be adjusted using the `linestyle` keyword:

```python
figure(figsize=(4, 3), dpi=80)

plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
```

### Adjusting the Plot: Axes Limits <a class="anchor" id="chapter2.2"></a>

Matplotlib does a decent job of choosing default axes limits for your plot, but sometimes it's nice to have finer control. The most basic way to adjust axis limits is to use the `plt.xlim()` and `plt.ylim()` methods:

```python
figure(figsize=(4, 3), dpi=80)

plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
```

We could use `xlim()` to reverse the order of axis:

```python
figure(figsize=(4, 3), dpi=80)

plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(1.5, -1.5);
```

We can also use the `plt.axis()` method to set the x and y limits with a single call, by passing a list which specifies [xmin, xmax, ymin, ymax]:

```python
figure(figsize=(4, 3), dpi=80)

plt.plot(x, np.sin(x))
plt.axis([-1, 10, -2, 2] )
```

### Labeling plots <a class="anchor" id="chapter2.3"></a>

```python
figure(figsize=(4, 3), dpi=80)
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)");
```

We can use `plt.legend()` and `label` to specify the legend of plot 

```python
figure(figsize=(4, 3), dpi=80)
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend();
```

In the object-oriented interface to plotting, rather than calling these functions individually, it is often more convenient to use the `ax.set()` method to set all these properties at once:


```python
figure(figsize=(4, 3), dpi=80)
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2),
       ylabel='x', xlabel='sin(x)',
       title='A Simple Plot');
```

### Scatter Plots with `plt.scatter` <a class="anchor" id="chapter3.2"></a>

```python
figure(figsize=(4, 3), dpi=80)
plt.scatter(x, y, marker='o');
```

```python
figure(figsize=(4, 3), dpi=80)
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.2,
            cmap='viridis')
plt.colorbar();
```

```python
from sklearn.datasets import load_iris
figure(figsize=(4, 3))
iris = load_iris()
features = iris.data.T

plt.scatter(features[0], features[1], alpha=0.2,
            s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```

## 4. Error bars <a class="anchor" id="chapter4"></a>

```python
figure(figsize=(4, 3))
x = np.linspace(0, 10, 50)
dy = 0.2
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x, y, yerr=dy, fmt='.k');
```

## 5. Histograms, Binnings, and Density <a class="anchor" id="chapter5"></a>

```python
figure(figsize=(6, 4))
data = np.random.randn(1000)
plt.hist(data, bins=50, alpha=0.5,
         histtype='stepfilled', color='steelblue',
         edgecolor='none');
```

```python
figure(figsize=(4, 3))
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

kwargs = dict(histtype='stepfilled', alpha=0.3, bins=40)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs);
```

```python
figure(figsize=(4, 3))
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T
plt.hist2d(x, y, bins=30, cmap='Blues')
cb = plt.colorbar()
cb.set_label('counts in bin')
```

## 6. Customizing Plot Legends <a class="anchor" id="chapter6"></a>

The simplest legend can be created with the `plt.legend()` command, which automatically creates a legend for any labeled plot elements:

```python
x = np.linspace(0, 10, 1000)
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(x, np.sin(x), '-b', label='Sinx')
ax.plot(x, np.cos(x), '--r', label='Cosinx')
ax.axis([0, 10, -2, 2])
leg = ax.legend(loc ='upper left', ncol=2);
```

## 8. Text and Annotation <a class="anchor" id="chapter8"></a>

The `ax.text` method takes an x position, a y position, a string, and then optional keywords specifying the color, size, style, alignment, and other properties of the text. ha is short for horizonal alignment. See the docstring of plt.text() and of mpl.text.Text() for more information on available options.

```python
x = np.linspace(0, 10, 1000)
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(x, np.sin(x), '-b', label='Sinx')
ax.axis([0, 10, -2, 2])
ax.text(1,np.sin(1),"Sin1", ha='center', size=10, color='grey')
ax.text(4,np.sin(4),"Sin4", ha='center', size=10, color='grey')
ax.text(6,np.sin(6),"Sin6", ha='center', size=10, color='grey')

leg = ax.legend(loc ='upper left', ncol=2);
```

`plt.annotate()` function creates some text and an arrow, and the arrows can be very flexibly specified.

```python
%matplotlib inline

fig, ax = plt.subplots()

x = np.linspace(0, 20, 1000)
ax.plot(x, np.cos(x))
ax.axis('equal')

#xytext: location of text, xy=(...): location for the point
ax.annotate('local maximum', xy=(6.28, 1), xytext=(8, 4),
            arrowprops=dict(facecolor='black', shrink=0.05))

ax.annotate('local minimum', xy=(5 * np.pi, -1), xytext=(2, -6),
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="angle3,angleA=0,angleB=-90"));
```
